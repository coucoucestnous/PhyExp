```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import glob
import re




path = "csv/"
filenames = glob.glob(path + "/*.csv")
filenames.sort(key=lambda f: int(re.sub('\D', '', f)))

N = len(filenames)
x0 = 400
dx = 215 # 20cm
nombredebandes = 6
histo = np.zeros((nombredebandes,N))


for file in filenames:
    xy = np.loadtxt(file, delimiter=",", skiprows=1, usecols=(1,2))
    plt.figure(figsize = (9,3))
    X = xy[:,0]
    Y = xy[:,1]
    plt.plot(X,Y, "+")
    plt.title(str(1) + "mn")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim(450,1650)
    plt.ylim(350,1000)
    plt.show()



for i in range(1,N+1):
    file = open("Batteur = 2.5/20°/CSV/%s.csv" % (i))
    xy = np.loadtxt(file, delimiter=",", skiprows=1, usecols=(1,2))
    X = xy[:,0]
    Y = xy[:,1]
    for j in range(nombredebandes):
        for k in range(len(X)):
            if x0 + j*dx <X[k]< x0 + (j+1)*dx:
                histo[j,i] += 1
            if X[k]< x0:
                histo[j,i] += 1
                print(X[k])
    histo = histo.T
    plt.figure(figsize = (9,3))
    plt.plot(X,Y, "+")
    plt.title(str(i))
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim(350,1650)
    plt.ylim(350,1000)
    plt.show()

```
